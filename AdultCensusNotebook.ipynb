```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
# import dataset
adult_dataset = pd.read_csv("datasets/Adult/adult.csv", header=0, index_col=None)

# replace and remove '?' values from the dataset
adult_dataset = adult_dataset.replace("?", np.nan)
adult_dataset.dropna(inplace=True)

# show the top 5 results from the dataset
adult_dataset.head()
```

```python
adult_dummies = pd.get_dummies(adult_dataset.drop(['income', 'education' ,'fnlwgt'], axis=1))
adult_income = adult_dataset['income'].apply(lambda x: 1 if x=='>50K' else 0)
```

```python
x_train, x_test, y_train, y_test = train_test_split(adult_dummies, adult_income, test_size=0.2)
```

```python
y_train.head()
```

```python
model = Sequential()
model.add(Dense(units=32, activation='relu', input_dim=len(x_train.columns)))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics='accuracy')
```

```python
model.fit(x_train, y_train, epochs=200, batch_size=32)
```

```python
y_predict = model.predict(x_test)
y_predict = [0 if val < 0.5 else 1 for val in y_predict]
```

```python
accuracy_score(y_test, y_predict)
```

```python
model.save('AdultCensusModel')
```
